```python
import pandas as pd
import matplotlib.pyplot as mpl
```

```python
data = pd.read_csv("data/items_per_order_aggregated.csv")
```

```python
data.head()
```

```python
mpl.bar(data['number_of_products_in_order'], data['count'])
mpl.title("Distribution of Order Sizes")
mpl.xlabel("Number of Products in Order")
mpl.ylabel("Number of Orders")
mpl.xticks(range(0,150,10))
mpl.xlim(0,150)
mpl.show()
```
